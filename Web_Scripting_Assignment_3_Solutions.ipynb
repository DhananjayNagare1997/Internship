```python
# scrape all product urls

product_urls = []

start = 0
end = 3

for page in range(start, end):  # for Loop for scrapping 3 pages
    urls = driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')

    for i in urls:
        # scraping urls
        product_urls.append(i.get_attribute("href"))

    # clicking the next button
    nxt_button = driver.find_element(By.XPATH, '//span[@class="s-pagination-strip"]/a')
    nxt_button.click()

    time.sleep(2)
```
